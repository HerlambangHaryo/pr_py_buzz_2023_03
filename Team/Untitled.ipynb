```python
# Import
import mysql.connector
import datetime

import pytz
utc=pytz.UTC 

# Connect to DB
host="localhost"
user="root" 
database="pr_mmbuzz_2022_06"
mydb = mysql.connector.connect(host=host,user=user,password="",database=database)
mycursor = mydb.cursor()

# Notifikasi Connect
datenow = datetime.datetime.now()
datenow_onlydate = datenow.strftime("%Y-%m-%d %H:%M:%S") 
latest_str = datenow.strftime("%Y-%m-%d") 
status_connect = "Connect at: " + str(datenow_onlydate)
print(status_connect)
```

```python
query = " Select teams_home_id   "
query += " , leagueapi_id "  
query += " , season  "  
query += " , league_country  "  
query += " , teams_home "   

query += " from football_fixtures "  
query += " where teams_home_id is not null  "  

query += " group by teams_home_id   "  
query += " , leagueapi_id   "  
query += " , season  "  
query += " , league_country  "  
query += " , teams_home "   

# query 1
mycursor = mydb.cursor()
mycursor.execute(query)
myresult = mycursor.fetchall()

print(len(myresult))

for x in myresult:   
    teams_home_id = x[0] 
    leagueapi_id = x[1] 
    season = x[2] 
    league_country = x[3] 
    teams_home = x[4]  
    
    query = " Select id   "
    query += " from teams "  
    query += " where teamapi_id = '"+str(teams_home_id)+"'   "  
    query += " and leagueapi_id = '"+str(leagueapi_id)+"'   "  
    query += " and season = '"+str(season)+"'   "  
    
    
    # query 1
    mycursor = mydb.cursor()
    mycursor.execute(query)
    myresult2 = mycursor.fetchall()
    
    if(len(myresult2) == 0):
        print("Kosong")
        
        insert = " INSERT INTO `teams`(   "
        insert += " `teamapi_id`, " 
        insert += " `leagueapi_id`, " 
        insert += " `season`, "  
        insert += " `name`, "  
        insert += " `country` "  
        insert += " ) VALUES ( "  
        insert += " '" + str(teams_home_id) +"', "  
        insert += " '" + str(leagueapi_id) +"', "  
        insert += " '" + str(season) +"', "  
        insert += " '" + str(teams_home.replace("'", "\\'")) +"', "  
        insert += " '" + str(league_country) +"' ) "  
        
        
        mycursor.execute(insert)
        mydb.commit()  
        
        word = "#" + str(teams_home_id)
        word += " " + teams_home
        word += " #" + str(leagueapi_id)
        word += " - " + str(season)
        word += " / " + str(league_country) 
        print(word) 
```
