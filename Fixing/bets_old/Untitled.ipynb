```python
def find_string_in_array(array, target_string):
    for item in array:
        if target_string in item:
            return item.split(target_string)[1]

# Example usage:
bils = [
    "/1_Home/match_winner_home,",
    "/1_Draw/match_winner_draw,",
    "/1_Away/match_winner_away,",
    "/2_Home/homeaway_home,",
    "/2_Away/homeaway_away,",
    "/3_Home/second_half_winner_home,",
    "/3_Draw/second_half_winner_draw,",
    "/3_Away/second_half_winner_away,"
]

target_string = "/1_Draw/"
result = find_string_in_array(bils, target_string)
print(result)  # Output: "match_winner_draw,"
```
