```python
import mysql.connector 
host="localhost"
user="root" 
database="pr_mmbuzz_2022_06"
mydb = mysql.connector.connect(host=host,user=user,password="",database=database)
mycursor = mydb.cursor()
# ------------------------------------------------------
query = "select "
query += " bet_id "
query += " , bet  "
query += " , value  "
query += " , columny  "
query += " , columnx  "
query += " , id  "
query += " from bets_old " 
query += " where bet_id IN (6) " 
query += " and value like '%over%' " 
query += " order by total desc " 
# ------------------------------------------------------   
# print(query)
# ------------------------------------------------------   
mycursor = mydb.cursor()
mycursor.execute(query)
result2 =  mycursor.fetchall()
# ------------------------------------------------------ 
mycursor.close()
mydb.close()
# ------------------------------------------------------  
counter = 0
pre_if = "if"
# ------------------------------------------------------  
for x in result2: 
    # -------------------------------------------------- 
    bet_id = str(x[0])
    bet = str(x[1])
    value = str(x[2])
    columny = str(x[3])
    columnx = str(x[4])
    idx = str(x[5])
    value_x = str(x[2]).replace(".", "").replace(" ", "_").lower()
    bet_x = bet_id +"_"+ str(x[1].replace(")", "").replace("(", "").replace("-", "_").replace("/", "").replace(" ", "").lower()) +"_"+ value_x 
    # --------------------------------------------------  
    if(counter != 0):
        pre_if = "elseif" 
    counter += 1
    # --------------------------------------------------   
    word = "@"+pre_if+"($pattern->"+columnx+"_perc >= 0.8)"
    print(word, flush=True)
    # --------------------------------------------------   
    word = "    <a"
    print(word, flush=True)
    # --------------------------------------------------   
    word = '        class="badge '+columny+'"'
    print(word, flush=True) 
    # --------------------------------------------------   
    word = '        target="_blank"'
    print(word, flush=True) 
    # --------------------------------------------------   
    route = "'Patterns.odd'"
    # --------------------------------------------------   
    word = '        href="{{route('+route+',['
    print(word, flush=True) 
    # --------------------------------------------------    
    word = "            'leagueapi_id' => $row->leagueapi_id,"
    print(word, flush=True)
    # --------------------------------------------------    
    word = "            'season' => $row->season,"
    print(word, flush=True)
    # --------------------------------------------------    
    word = "            'fixtureapi_id' => $row->fixtureapi_id,"
    print(word, flush=True)
    # --------------------------------------------------    
    word = "            'odd' => '"+bet_x+"',"
    print(word, flush=True)
    # --------------------------------------------------    
    word = '        ])}}"'
    print(word, flush=True)
    # --------------------------------------------------   
    word = "    >"
    print(word, flush=True) 
    # --------------------------------------------------   
    word =  "        " + value + " / "
    print(word, flush=True)
    # --------------------------------------------------   
    word =  "        @if(!is_null($row->end_"+columnx+"))" 
    print(word, flush=True)
    # --------------------------------------------------   
    word =  "            {{$row->end_"+columnx+"}}" 
    print(word, flush=True)
    # --------------------------------------------------   
    word =  "        @elseif(!is_null($row->pre_"+columnx+"))" 
    print(word, flush=True)
    # --------------------------------------------------   
    word =  "            {{$row->pre_"+columnx+"}}" 
    print(word, flush=True)
    # --------------------------------------------------   
    word =  "        @endif" 
    print(word, flush=True)
    # --------------------------------------------------   
    word =  "        => {{ $pattern->"+columnx+"_perc * 100 }}%"
    print(word, flush=True)
    # --------------------------------------------------   
    word = "    </a>"
    print(word, flush=True)
    # --------------------------------------------------     
word = "@endif"
print(word, flush=True)
# --------------------------------------------------   
```
