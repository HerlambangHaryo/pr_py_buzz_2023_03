```python
from a_settings.messages import *  
from a_models.football_odds import *  

def get_date(space):  
    # ----------------------------------------------------------  
    space += "__"
    # ----------------------------------------------------------  
    print(space + "get_date()")
    # ----------------------------------------------------------  
    today = datetime.datetime.now()  
    today_day = today.strftime("%Y-%m-%d")   
    # ----------------------------------------------------------  
    DD = datetime.timedelta(days=days0)
    week = today - DD
    tgl_awal = week.strftime("%Y-%m-%d") 
    # ----------------------------------------------------------  
    DD = datetime.timedelta(days=days1)
    yesterday = today - DD
    tgl_akhir = yesterday.strftime("%Y-%m-%d")  
    # ----------------------------------------------------------  
    space += "__"
    # ----------------------------------------------------------  
    print(space + "today : " + str(today_day))
    print(space + "days : " + str(days0))
    print(space + "days : " + str(days1))
    print(space + "tgl_awal : " + tgl_awal)
    print(space + "tgl_akhir : " + tgl_akhir) 
    # ---------------------------------------------------------- 
    fO_get_group_league_status_match_finished(tgl_awal, tgl_akhir, today_day, space)
    # ---------------------------------------------------------- 
    
if __name__ == '__main__': 
    # --------------------------------------------------  
    its_start()
    # --------------------------------------------------  
    space = "  __"
    print(space + "main()") 
    # --------------------------------------------------   
    get_date(space)
    # -------------------------------------------------- 
    its_end()  
    ####################################################
```
