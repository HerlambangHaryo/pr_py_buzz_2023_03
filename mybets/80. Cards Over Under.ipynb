```python
import mysql.connector 
# ------------------------------------------------------
host="localhost"
user="root" 
database="pr_mmbuzz_2022_06"
mydb = mysql.connector.connect(host=host,user=user,password="",database=database)
mycursor = mydb.cursor()
# ------------------------------------------------------
query = "select "
query += " bet_id "
query += " , bet  "
query += " , value  "
query += " , columny  "
query += " , columnx  "
query += " from bets_old " 
query += " where bet_id = " + INIT_bet_id   
query += " order by bet_id, total asc "   
# ------------------------------------------------------   
print(query)
# ------------------------------------------------------   
mycursor = mydb.cursor()
mycursor.execute(query)
result2 =  mycursor.fetchall()
# ------------------------------------------------------    
total_rows = len(result2)
# ------------------------------------------------------    
mycursor.close()
mydb.close() 
```

```python
# ----------------------------------------------------- 
space = "    "
# ----------------------------------------------------- 
for x in result2:  
    # ------------------------------------------------- 
    bet_id = str(x[0])
    bet = str(x[1])
    value = str(x[2])
    columny = str(x[3])
    columnx = str(x[4])
    # -------------------------------------------------  
    # -------------------------------------------------  
    if 'Over' in value:
        # @if(!is_null($row->pre_asian_handicap_over_plus_65) && !is_null($row->pre_asian_handicap_Under_plus_65) )
        # ---------------------------------------------  
        word = "@if(!is_null($row->pre_"+columnx+") && !is_null($row->pre_"+columnx.replace("_over_", "_under_")+") )" 
        print(word, flush=True)
        # @if( ($row->pre_asian_handicap_over_plus_65 + $row->pre_asian_handicap_Under_plus_65) < 4 )
         # ---------------------------------------------  
        word = space + "@if( ($row->pre_"+columnx+" + $row->pre_"+columnx.replace("_over_", "_under_")+") < 4 )" 
        print(word, flush=True)
        # ---------------------------------------------  
        word = space + space + '<div class="btn-group mt-2">' 
        print(word, flush=True)
        # ---------------------------------------------
    # -------------------------------------------------   
    # -------------------------------------------------   
    word = space + space + space + '<a' 
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + '@if(!is_null($row->pre_'+columnx+'))'
    print(word, flush=True)
    # ------------------------------------------------- 
    route = "{{route('Mybets.datacreateanalysis',["
    # ------------------------------------------------- 
    word = space + space + space + space + space + 'href="'+route
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + space + space + "'leagueapi_id' => $row->leagueapi_id,"
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + space + space + "'season' => $row->season,"
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + space + space + "'fixtureapi_id' => $row->fixtureapi_id,"
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + space + space + "'betsapi_id' => "+INIT_bet_id+","
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + space + space + "'value' => " + "'" + value + "'"+ ","
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + space + space + "'odd' => $row->pre_"+columnx+","
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + space + '])}}"'
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + '@endif'
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + 'target="_blank"'
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + 'class="btn bg-gray-700 btn-sm text-white">'
    print(word, flush=True) 
    # -------------------------------------------------   
    if 'Over' in value:
        prep_val = "ou"
    elif 'Under' in value:
        prep_val = value.replace("Under ", "")
    # -------------------------------------------------   
    word = space + space + space + space + space +  prep_val
    print(word, flush=True)
    # -------------------------------------------------  
    word = space + space + space + space + space + '<br/>'
    print(word, flush=True)
    # ------------------------------------------------- 
    word = space + space + space + space + space + "{{ $row->pre_"+columnx+" }}"
    print(word, flush=True)
    # -------------------------------------------------  
    word = space + space + space + '</a>'
    print(word, flush=True)
    # -------------------------------------------------   
    # -------------------------------------------------  
    if 'Under' in value:
        # ---------------------------------------------
        word = space + space + '</div>'
        print(word, flush=True)
        # ---------------------------------------------
        word = space + '@endif'
        print(word, flush=True)
        # ---------------------------------------------
        word = '@endif'
        print(word, flush=True)
        # ---------------------------------------------
        word = ''
        print(word, flush=True)
        # ---------------------------------------------
    # -------------------------------------------------  
    # -------------------------------------------------  
```
